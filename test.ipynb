```python
import numpy as np
import timeit

# Create two large matrices
A = np.random.rand(10, 784)
B = np.random.rand(784, 1)

# Benchmark np.matmul
time_taken = timeit.timeit(lambda: np.matmul(A, B), number=60_000)
print(f"Time taken for 10 matrix multiplications: {time_taken:.6f} seconds")
```

```python
import time
for i in range(1,150):
    a = np.ones(shape=i*100_000)
    b = np.ones(shape=i*100_000)
    start_time = time.time()
    x = a*b
    end_time = time.time()
    timeArr.append(end_time-start_time)
```
